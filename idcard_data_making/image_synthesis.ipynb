```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
from google.colab.patches import cv2_imshow

# 파일에서 정보 읽기
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return [line.strip() for line in f.readlines()]

# 각각의 텍스트 파일로부터 정보를 읽어옴
names_kor = read_file('names_korean.txt')      # 한글 이름 파일
names_hanja = read_file('names_hanja.txt')  # 한자 이름 파일
ssns = read_file('resident_numbers.txt')                # 주민등록번호 파일
addresses = read_file('addresses.txt')       # 주소 파일
issue_dates = read_file('issue_dates.txt')  # 발급일자 파일
issuers = read_file('issue_authorities.txt')          # 발급기관 파일

# 정보 세트를 4개의 세트로 구성
info_sets = []
for i in range(len(names_kor)):
    info_set = {
        "name_kor": names_kor[i],
        "name_hanja": names_hanja[i],
        "ssn": ssns[i],
        "address": addresses[i],
        "issue_date": issue_dates[i],
        "issuer": issuers[i]
    }
    info_sets.append(info_set)

# 이미지 생성 및 합성 함수 정의
def generate_id_card_with_face(info_set, face_image_path, output_path):
    # 주민등록증 기본 이미지 불러오기 (OpenCV를 사용하여 이미지 로드)
    image = cv2.imread('idcard_blank.png')

    # OpenCV 이미지를 PIL 이미지로 변환
    image_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # 폰트 설정 (경로는 사용 환경에 맞게 수정)
    font_path = "/content/HY견고딕.TTF"
    font_title = ImageFont.truetype(font_path, 45)  # 제목 폰트 사이즈
    font_name = ImageFont.truetype(font_path, 35)  # 다른 텍스트 폰트 사이즈
    font_ssn = ImageFont.truetype(font_path, 30)
    font_address = ImageFont.truetype(font_path, 25)
    font_date = ImageFont.truetype(font_path, 25)
    font_issuer = ImageFont.truetype(font_path, 30)

    # 텍스트를 개별 글자별로 추가하기 위해 글자 간격 설정
    def draw_text_with_spacing(draw, position, text, font, fill=(0, 0, 0), spacing=0):
        x, y = position
        for char in text:
            draw.text((x, y), char, font=font, fill=fill)
            x += font.getsize(char)[0] + spacing  # 각 글자의 너비에 추가적인 간격을 더함

    def draw_text_wrapped(draw, text, position, font, max_width, spacingline, spacing=0):
        x, y = position
        lines = []
        words = text.split(' ')
        line = ""

        for word in words:
            test_line = line + word + " "
            width, _ = draw.textsize(test_line, font=font)
            if width <= max_width:
                line = test_line
            else:
                lines.append(line)
                line = word + " "

        lines.append(line)  # 마지막 라인 추가
        nx = x
        for line in lines:
            x = nx
            for char in line:
                draw.text((x, y), char, font=font, fill=(0, 0, 0))
                x += font.getsize(char)[0] + spacing  # 각 글자의 너비에 추가적인 간격을 더함.
            y += font.getsize(line)[1] + spacingline

    max_width = 350  # 줄바꿈을 위한 최대 너비 설정

    # 텍스트 좌표 설정
    coordinates = {
        "title": (110, 50),  # 주민등록증 제목
        "name_kor": (83, 130),  # 이름 (한글)
        "name_hanja": (210, 130),  # 이름 (한자)
        "ssn": (70, 190),  # 주민등록번호
        "address": (60, 250),  # 주소
        "issue_date": (250, 350),  # 발급일자
        "issuer": (155, 380)  # 발급기관
    }

    # 텍스트를 이미지에 추가
    draw = ImageDraw.Draw(image_pil)
    draw_text_with_spacing(draw, coordinates["title"], "주민등록증", font=font_title, spacing=7)
    draw_text_with_spacing(draw, coordinates["name_kor"], info_set["name_kor"], font=font_name, spacing=6)
    draw_text_with_spacing(draw, coordinates["name_hanja"], info_set["name_hanja"], font=font_name, spacing=1)
    draw_text_with_spacing(draw, coordinates["ssn"], info_set["ssn"], font=font_ssn, spacing=2)
    draw_text_wrapped(draw, info_set["address"], coordinates["address"], font_address, max_width, spacingline=5, spacing=3)
    draw_text_with_spacing(draw, coordinates["issue_date"], info_set["issue_date"], font=font_date, spacing=1)
    draw_text_with_spacing(draw, coordinates["issuer"], info_set["issuer"], font=font_issuer, spacing=5)


    # 얼굴 이미지 불러오기 및 크기 조정
    face_image = Image.open(face_image_path)
    face_width, face_height = 230, 280  # 얼굴 이미지 크기 (픽셀 단위)
    face_image = face_image.resize((face_width, face_height), Image.ANTIALIAS)

    # 얼굴 이미지의 합성 위치 설정 (좌측 상단 기준)
    position = (472, 40)  # (x, y) 좌표로 설정

    # 얼굴 이미지를 주민등록증 이미지에 합성
    image_pil.paste(face_image, position, face_image.convert('RGBA'))

    # 결과를 다시 OpenCV 이미지로 변환
    image = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

    # 최종 이미지 저장 및 출력
    cv2.imwrite(output_path, image)
    cv2_imshow(image)

# 각 정보 세트에 대해 이미지 생성 및 출력
face_image_paths = ['seed22.png', 'seed23.png', 'seed24.png', 'seed25.png']  # 얼굴 이미지 경로 리스트

for idx, info_set in enumerate(info_sets):
    output_path = f'final_image_{idx+1}.png'
    generate_id_card_with_face(info_set, face_image_paths[idx], output_path)
```
